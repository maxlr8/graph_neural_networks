## __Basic Graph Network - PyTorch__

```python
# Define a Graph.
# Store it in the format of a torch tensor and graph representation.

import torch
import networkx as nx

from torch_geometric.data import Data
from torch_geometric.utils.convert import to_networkx
```

```python
# Creating edge lists

edge_list = torch.tensor(
    [[0, 0, 0, 1, 2, 2, 3, 3],   # Source Nodes
    [1, 2, 3, 0, 0, 3, 2, 0]],  # Target Nodes
    dtype = torch.long
)

# Creating 6 Feature Tensors for every node.

node_features = torch.tensor(
    [[-8, 1, 5, 8, 2, -3],  # Node 1
    [-1, 0, 2, -3, 0, 1],  # Node 2
    [1, -1, 0, -1, 2, 1],  # Node 3
    [0, 1, 4, -2, 3, 4]],  # Node 4
    dtype=torch.long
)

# Creating 1 Weight for each edge

edge_weights = torch.tensor(
    [[35.],  # Weight for the edge (0,1)
    [48.],  # Weight for the edge (0,2)
    [12.],  # Weight for the edge (0,3)
    [10.],  # Weight for the edge (1,0)
    [70.],  # Weight for the edge (2,0)
    [5.],   # Weight for the edge (2,3)
    [15.],  # Weight for the edge (3,2)
    [8.]]   # Weight for the edge (3,0)
)
```

```python
# Make a data object to store Graph Information

data = Data(x=node_features, edge_index=edge_list, edge_attr=edge_weights)

# Print the graph information

print("Number of Nodes: ", data.num_nodes)
print("Number of Edges: ", data.num_edges)
print("Number of features per Node (Length of feature vector): ", data.num_node_features)
print("Number of features per Edge (Length of Edge vector): ",data.num_edge_features)
```

```python
#Plot the graph

G = to_networkx(data)
nx.draw_networkx(G)
```
